```python
import numpy as np

class RelationshipEnvironment:
    def __init__(self):
        # Define the state space
        self.relationship_status = ['single', 'dating', 'in a relationship', 'married']
        self.emotional_wellbeing = ['low', 'medium', 'high']
        
        # Define the action space
        self.actions = ['go to date', 'focus on self-improvement', 'end a relationship', 'apologize and reconcile', 'build connection']
        
        # Initialize the state
        self.state = [0, 1]  # Initial state: single and medium emotional wellbeing
        
    def reset(self):
        self.state = [0, 1]  # Reset to initial state
        return self.state
    
    def step(self, action):
        relationship_status, emotional_wellbeing = self.state
        
        # Define rewards
        rewards = {
            'find compatible partner': 10,
            'improve emotional state': 5,
            'experience rejection or heartbreak': -10,
            'stay single': 0
        }
        
        # Define the state transition logic
        if action == 'go to date':
            if relationship_status == 0:  # single
                relationship_status = 1  # dating
                reward = rewards['find compatible partner']
            else:
                reward = rewards['experience rejection or heartbreak']
        
        elif action == 'focus on self-improvement':
            if emotional_wellbeing < 2:  # not high
                emotional_wellbeing += 1
                reward = rewards['improve emotional state']
            else:
                reward = rewards['stay single']
        
        elif action == 'end a relationship':
            if relationship_status > 0:  # not single
                relationship_status = 0  # single
                reward = rewards['stay single']
            else:
                reward = rewards['experience rejection or heartbreak']
        
        elif action == 'apologize and reconcile':
            if relationship_status > 1:  # in a relationship or married
                emotional_wellbeing = 2  # high
                reward = rewards['improve emotional state']
            else:
                reward = rewards['experience rejection or heartbreak']
        
        elif action == 'build connection':
            if relationship_status > 0 and emotional_wellbeing < 2:  # dating or in a relationship and not high emotional wellbeing
                emotional_wellbeing += 1
                reward = rewards['improve emotional state']
            else:
                reward = rewards['stay single']
        
        # Update the state
        self.state = [relationship_status, emotional_wellbeing]
        
        return self.state, reward

# Example usage
env = RelationshipEnvironment()
state = env.reset()
print("Initial state:", state)

action = 'go to date'
new_state, reward = env.step(action)
print("Action:", action)
print("New state:", new_state)
print("Reward:", reward)
```
